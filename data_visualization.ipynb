```python
t=np.arange(0.0,24,0.01)

s=np.sin(t)
fig,ax=plt.subplots()
ax.plot(t,s)

s=np.cos(t)
fig,ax=plt.subplots()
ax.plot(t,s)

t=np.arange(0.0,1,0.01)
s=np.tan(t)
fig,ax=plt.subplots()
ax.plot(t,s)
ax.set(xlabel='radians',ylabel='tan(x)',title='Tangent plot')
ax.grid()
```
