```python
df3=pd.read_csv("assignment_aums_2.csv")
df3
```

```python
df3.info()
```

```python
df3["actual_year"]=pd.to_datetime(df3["Year"],format='%m/%d/%y %H:%M')
df3["YEAR"]=df3["actual_year"].dt.year
df3.drop(columns="actual_year",inplace=True)
df3
```

```python
df3[df3["YEAR"]==2014]
```

```python
df3.groupby("Educational Attainment")["Population Count"].sum()
```

```python
df3[(df3["Age"] != "00 to 17")].groupby("Educational Attainment")[["Population Count"]].sum()/df3.groupby("Educational Attainment")[["Population Count"]].sum()*100
```

```python
percentage_distribution = (df3.query("Age != '00 to 17'").groupby("Educational Attainment")[["Population Count"]].sum() / df3.groupby("Educational Attainment")[["Population Count"]].sum()) * 100
percentage_distribution
```

```python
df3.info()
```

```python
df3[["l_age","u_age"]]=df3["Age"].str.extract(r'(\d+) to (\d+|\d+\+)')
df3
```

```python
import matplotlib.pyplot as plt

# Sample data
x1 = df3[df3["Educational Attainment"]=="No high school diploma"]["Personal Income"]
x2 = df3[df3["Educational Attainment"]=="Bachelor's degree or higher"]["Personal Income"]
values = df3["Personal Income"]


# Create a bar graph
plt.bar(x1,values)

# Add labels and a title
plt.xlabel('Categories')
plt.ylabel('Personal Income')
plt.title('Bar Graph ')

# Display the graph
plt.show()
```
