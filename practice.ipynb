```python
df=df[df['AGE']<999]
df=df[df['SEX']<2]
```

```python
ages=df['AGE']
ages
```

```python
df_small=df.groupby(['AGE','SEX'],group_keys=False).sum()
```

```python
df_small
```

```python
for i in range(0,len(ages),2):
    for j in range(2,8):
        df_small.iloc[(i,j)]=df_small.iloc[(i,j)]/df_small.iloc[(i+1,j)]
df_small
```

```python
a=df_small.index
v=[]
for i in a:
    v.append(i[1])
v=np.array(v)
df_small['SEXX']=v
df_small
```

```python
df_small=df_small[df_small['SEXX']<1]
df_small
```

```python
a=np.unique(ages)
df_small['AGEE']=a
df_small
```

```python
plt.plot(df_small['AGEE'],df_small['POPESTIMATE2010'],label="2010")
plt.plot(df_small['AGEE'],df_small['POPESTIMATE2011'],label="2011")
plt.plot(df_small['AGEE'],df_small['POPESTIMATE2012'],label="2012")
plt.plot(df_small['AGEE'],df_small['POPESTIMATE2013'],label="2013")
plt.plot(df_small['AGEE'],df_small['POPESTIMATE2014'],label="2014")
plt.plot(df_small['AGEE'],df_small['POPESTIMATE2015'],label="2015")
plt.legend()
```

```python
df2=pd.read_csv("educ_inc.csv")
df2
```

```python
cat = df2['Educational Attainment']
mean_income = df2['Personal Income']

bar_width=0.35
index=np.arange(len(cat))
plt.barh(index-bar_width / 2,mean_income,bar_width,label='Mean Income',color='blue')
plt.barh(index+bar_width / 2,mean_income,bar_width,label='Mean Income',color='green')

plt.xlabel("Mean Income")
plt.yticks(index,cat)
plt.title("Personal Income Distribution by Education Level in California")

plt.legend(['No diploma','bachelors or highers'])
```
