Faulty dataset preprocessing

```python
df = pd.read_csv("Faulty_dataset.csv")
df
```

```python
df.isnull().sum()
```

```python
def fillNames(row):
    if pd.isnull(row['name']):
        if row['gender']=="Male":
            return "John Doe"
        elif row['gender']=='Female':
            return "Jane Doe"
    return row['name']

df['name']=df.apply(fillNames,axis=1)
df
```

```python
lb=0
ub=100
df['10th']=df['10th'].apply(lambda x:lb if x<lb else(ub if x>ub else x))
df
```
