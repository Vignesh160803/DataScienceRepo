Note that the first row of the percent_change column will be NaN, since there is no previous row to compare it to.

11.	Plot the gender ratio versus age.

```python
df=pd.read_csv("nc-est2015-agesex-res.csv")
df
```

```python
df=df[df["AGE"]<999]
df=df[df["SEX"]<2]
```

```python
age=df["AGE"]
age
```

```python
d=df.groupby(["AGE","SEX"]).sum()
```

```python
d
```

```python
for i in range(0,len(age),2):
    for j in range(2,8):
        d.iloc[(i,j)]=d.iloc[(i,j)]/d.iloc[(i+1,j)]
d
```

```python
a=d.index
v=[]
for i in a:
    v.append(i[1])
v=np.array(v)
d["SEXX"]=v
```

```python
d
```

```python
d=d[d["SEXX"]<1]
d
```

```python
a=np.unique(age)
d["AGEE"]=a
d
```

```python
plt.plot(d["AGEE"],d["POPESTIMATE2010"],label="2010")
plt.plot(d["AGEE"],d["POPESTIMATE2011"],label="2011")
plt.plot(d["AGEE"],d["POPESTIMATE2012"],label="2012")
plt.plot(d["AGEE"],d["POPESTIMATE2013"],label="2013")
plt.plot(d["AGEE"],d["POPESTIMATE2014"],label="2014")
plt.plot(d["AGEE"],d["POPESTIMATE2015"],label="2015")
plt.legend()
```
